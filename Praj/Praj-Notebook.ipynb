```python
#Gets the population values in 1960 (but of course it's gotta be a string, so it's more complicated)
string_nums=population_density['Pop. 1960']
#headers=[cols[0],cols[4]]
sixties_pops=pd.Series(neighborhoods)

#Creates empty List to hold int values of sixties
sixties_ints=[]

#Converts string_nums string values into integers and stores them into sixties_pops[]
for x in string_nums:
    temp=int(x.replace(',', ''))
    sixties_ints.append(temp)

sixties_ints=pd.Series(sixties_ints)
sixties_pops=pd.concat([neighborhoods,sixties_ints],axis=1)
sixties_pops.columns=['Neighborhood','Pop. 1960']


sixties_pops=sixties_pops.sort_values('Pop. 1960',ascending=False)

###Credit goes to https://www.youtube.com/watch?v=nKxLfUrkLE8 for helping me to setup the bar chart below
# Puts the Neighborhoods into a list called 'places'.
# Must do this individually for each population because of ordering
places=[]
for x in sixties_pops['Neighborhood']:
    places.append(x)

#Creates a bar chart
bar=sixties_pops.plot(kind='bar',figsize=(20,10),xlabel='Neighborhoods',ylabel='Population',title='Population in 2010')
x_indexes= np.arange(len(places))
plt.xticks(ticks=x_indexes,labels=places,fontsize=10)
plt.show()
```

```python
#Gets the population values in 1960 (but of course it's gotta be a string, so it's more complicated)
string_nums=population_density['Persons / sq. mi (2010)']
#headers=[cols[0],cols[4]]
pop_densities=pd.Series(neighborhoods)

#Creates empty List to hold int values of sixties
pop_density_vals=[]

#Converts string_nums string values into integers and stores them into sixties_pops[]
for x in string_nums:
    temp=float(x.replace(',', ''))
    pop_density_vals.append(temp)

pop_density_vals=pd.Series(pop_density_vals)
pop_densities=pd.concat([neighborhoods,pop_density_vals],axis=1)
pop_densities.columns=['Neighborhood','Persons / sq. mi (2010)']


pop_densities=pop_densities.sort_values('Persons / sq. mi (2010)',ascending=False)

###Credit goes to https://www.youtube.com/watch?v=nKxLfUrkLE8 for helping me to setup the bar chart below
# Puts the Neighborhoods into a list called 'places'.
# Must do this individually for each population because of ordering
places=[]
for x in pop_densities['Neighborhood']:
    places.append(x)

#Creates a bar chart
bar=pop_densities.plot(kind='bar',figsize=(20,10),xlabel='Neighborhoods',ylabel='Population',title='Persons / sq. mi (2010)')
x_indexes= np.arange(len(places))
plt.xticks(ticks=x_indexes,labels=places,fontsize=10)
plt.show()
```
